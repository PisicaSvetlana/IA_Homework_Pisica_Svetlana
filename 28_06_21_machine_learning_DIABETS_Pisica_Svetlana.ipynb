```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('diabetes.csv')          # in variabila df se incarca datele din fisierul cu date de analizat
print ('Datele din fisier toate coloanele \n\n', df)   # afisam datele din var df cu datele din fisier cu datele pe care le analizam
                                       # ultima coloana este coloana cu rezultatele - target 
    
 ###########################################################################################################

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 


df = df.sample (frac=1)        #am amestecat rindurile din tabel  ca sa nu avem intii 1111 apoi 000 la target, 
print ('\n\n Datele din fisier toate coloanele  dupa amestecare\n\n ', df)    # trebuie sa fie amestecate pentru a putea analiza corect '''


#####################################################################################

# X  trebuie sa fie mare 

X = df.drop(['Outcome'], axis = 1 )    # scoatem din tabel ultima coloana, cea cu rezultatele - target

print ('\n\n Datele din fisier / tabel fara ultima coloana OUTCOME \n\n', X)      # afisam tabelul fara ultima coloana - target 


#####################################################################################

y = df['Outcome'].values
print ('\n\n Valorile din ultima coloana OUTCOME puse in y \n\n', y)


######################################################################

X_train, X_test, y_train, y_test = train_test_split (X, y, random_state=42)   # antrenam, testam 
knn = KNeighborsClassifier()
knn.fit (X_train, y_train)


y_pred = knn.predict (X_test)                # calcularea prezicerii
print ( '\n Prezicerea lui y \n\n', y_pred)   # afisarea prezicerii
print ('\n Acuratetea dupa prima incercare =', accuracy_score(y_pred, y_test))    # calcularea si afisarea acuratetii 



from sklearn.preprocessing import StandardScaler

std = StandardScaler ()
X_train_scaled = std.fit_transform(X_train)
X_test_scaled = std.transform (X_test)
knn.fit (X_train_scaled, y_train)


y_pred = knn.predict (X_test_scaled)     # calcularea prezicerii a 2 oara

print ('\n Prezicerea lui y \n \n ', y_pred )     # afisarea prezicerii 

print ('\n Acuratetea dupa  a doua incercare =', accuracy_score(y_pred, y_test) )   # aclculare si afisare acuratete a 2 oara 

```
